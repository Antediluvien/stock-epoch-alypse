```python
import pandas as pd
```

```python
csv_path = "EOD_20181027_v1.csv"
# col_names = ["Stock", "Date","Adj Close"]
```

```python
df = pd.read_csv(csv_path)
```

```python
df.columns = ['adj_close','stock','date','records']
```

```python
stock_df = df[['date','adj_close','stock']]
```

```python
stock_df.head()
```

```python
stock_df.dtypes
```

```python
stock_df['date']=pd.to_datetime(df.date)
```

```python
pivot_df = stock_df.pivot(index='date', columns='stock', values='adj_close')
```

```python
pivot_df = pivot_df.reset_index()
```

```python
pivot_df.sort_values(by=['date'],inplace=True)
```

```python
pivot_df.tail(1000)
```

```python
pivot_df.to_csv("stocks_clean.csv", index=False)
```
